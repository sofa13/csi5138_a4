# Setup

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib import rcParams

%matplotlib inline

from IPython.display import Image
```

```python
def plotsidebyside(path1, path2, fig1, fig2, title1=None, title2=None):
    # figure size in inches optional
    rcParams['figure.figsize'] = fig1, fig2

    # read images
    img_A = mpimg.imread(path1)
    img_B = mpimg.imread(path2)

    # display images
    fig, ax = plt.subplots(1,2)
    
    if title1:
        ax[0].set_title(title1)
    ax[0].axis('off')
    ax[0].imshow(img_A)
    
    if title2:
        ax[1].set_title(title2)
    ax[1].axis('off')
    ax[1].imshow(img_B)    
```

# VAE-MNIST

### Varying latent dimension

While keeping intermediate dimension fixed at 512, we tried latent dimensions 2, 5, and 10. We did not find this sped up training, since it took about 30 seconds per epoch regardless of the latent dimensions we tried.

#### Loss

Overall we found the loss decreases from latent 2 to 5 to 10, where we've plotted the loss with latent dimension 2 and 10 below

```python
plotsidebyside('./images/vae_mnist_loss_latent2_intermediate512.png', './images/vae_mnist_loss_latent10_intermediate512.png', 16, 11, title1="Latent_dim: 2, Intermediate_dim: 512", title2="Latent_dim: 10, Intermediate_dim: 512")
```

#### Gallery

Overall we find higher latent dimensions yields better quality generated images.

```python
plotsidebyside('./images/vae_mnist_gallery_latent2_intermediate512.png', './images/vae_mnist_gallery_latent10_intermediate512.png', 16, 11, title1="Latent_dim: 2, Intermediate_dim: 512", title2="Latent_dim: 10, Intermediate_dim: 512")
```

#### Latent space

Although the latent dimension 10 gives better quality generated images2, one handy thing about having a latent dimension 2 is that all the information is condensed into a space that is easily plotted. Below is a plot of the latent space of each digit, with latent dimension 2 and intermediate dimension 512. We can see digits 0 and 1, and 7 are more separated than the other digits.

```python
Image(filename='./images/vae_mnist_latentplot_latent2_intermediate512.png')
```

### Varying model complexity

While keeping latent dimension fixed at 2, we tried varying the model complexity by decreasing/increasing the intermediate dimension to 256/1024 respectively. Not surprisingly the calculations were fast for intermediate dimension 256, taking on average about 15s per epoch. In contrast intermediate dimension 512 took 30s per epoch and 1024 took about 60s per epoch.

#### Loss

Overall we find varying the intermediate dimension, with a larger dimension of 1024 we reduce the loss more quickly, but with all intermediate dimensions 256, 512, and 1024, the loss between the training and test start to diverge around loss 155. Note the loss for intermediate dimension 512 is shown above.

```python
plotsidebyside('./images/vae_mnist_loss_latent2_intermediate256.png', './images/vae_mnist_loss_latent2_intermediate1024.png', 16, 11, title1="Latent_dim: 2, Intermediate_dim: 256", title2="Latent_dim: 2, Intermediate_dim: 1024")
```

#### Gallery

Overall we find changing the intermediate dimensions does not yield better quality generated images.

```python
plotsidebyside('./images/vae_mnist_gallery_latent2_intermediate256.png', './images/vae_mnist_gallery_latent2_intermediate1024.png', 16, 11, title1="Latent_dim: 2, Intermediate_dim: 256", title2="Latent_dim: 2, Intermediate_dim: 1024")
```
